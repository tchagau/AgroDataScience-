```python
def ccaracterising(x, moy, a=.5,b=1,c=1.5):
    
    a, b, c = moy*a, moy*b, moy*c
    
    if   (a >= x): return 0
    elif (a <= x) & (x < b): return 1
    elif (b <= x) & (x < c): return 2
    elif (c <= x): return 3
    return -1

# Carterisation du capital
mb['capital6c'] = mb[capitaux].apply(sum, axis=1).apply(lambda r: ccaracterising(r, mb[capitaux].apply(sum, axis=1).mean()))
mb['capital6c'].value_counts().plot(kind='pie', autopct='%1.0f%%')
```

```python
def rcaracterising(r):
    list_vars = ['Revenu_monétaire_Cult._Pluviale','Revenu_monétaire_Cult._Irriguée',
                'Revenu_monétaire_Cult._Décrue','Sous-total_Revenu_monétaire_production_animale','Sous-Total_Revenu_monétaire_non_agricle']
    winner = None
    for col in list_vars:
        #print(1.*(r[col] >=0.3*r['Total_Revenu_monétaire']))
        if r[col] >= 0.30*r['Total_Revenu_monétaire']:
            #print(col)
            winner = col
            break
    if winner == None: return -1
    for (i,k) in enumerate(list_vars):
        #print(i, k)
        if k==col: return i
    return 'Bug'
'''
Total_Revenu_monétaire
Revenu_monétaire_Cult._Pluviale
Revenu_monétaire_Cult._Irriguée
Revenu_monétaire_Cult._Décrue
Sous-total_Revenu_monétaire_production_animale
Sous-Total_Revenu_monétaire_non_agricle
'''

# Carterisation du revenu
mb['revenu6c'] = mb.apply(lambda r: rcaracterising(r), axis=1)
mb['revenu6c'].value_counts()
capitalrevenu = ['capital6c','revenu6c']
```
